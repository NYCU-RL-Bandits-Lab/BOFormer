```python
import numpy as np
from matplotlib import pyplot as plt
import pickle
import glob
datasets = ["BC"]
algs = ["./BOFormer_testings/model0/",
        "./BOFormer_testings/model4/",
        "./BOFormer_testings/model8/",
        "./BOFormer_testings/model9/"]
algs_name = ["GP(0.1-0.4), BC=0.5, lr=10e-05, npos=31",
        "GP(0.4-0.8), BC=0.1, lr=10e-06, npos=31",
        "GP(0.4-0.8), BC=0.1, lr=10e-05, npos=31",
        "GP(0.4-0.8), BC=1.0, lr=10e-05, npos=61"]
T = 100
index = 0
path = ""
history_list = dict()
for d in datasets:
    history_list[d] = dict()
    for i in range(len(algs)): 
        history_list[d][algs_name[i]] = []
        print(algs_name[i], len(glob.glob(path +  algs[i] + "*_" + d + '*.pkl')))
        for file in glob.glob(path +  algs[i] + "*_" + d + '*.pkl'):
            with open(file, 'rb') as f:
                history = np.load(f, allow_pickle=True)
                history_list[d][algs_name[i]].append(history)

    
# with open('mobo_regret.pickle', 'wb') as handle:
#     pickle.dump(regrets, handle, protocol=pickle.HIGHEST_PROTOCOL)

```

```python
for d in datasets:
    for i in range(len(algs)):
        regret = [history['hypervolume_true'] for history in history_list[d][algs_name[i]]]
        regret = np.mean(regret,axis=0) 
        plt.plot(regret, label=algs_name[i])
    plt.title(d)
    plt.legend()
```

```python
for d in datasets:
    for i in range(len(algs)):
        regret = [history['hypervolume_true'] for history in history_list[d][algs_name[i]]]
        regret = np.std(regret,axis=0) 
        plt.plot(regret, label=algs_name[i])
    plt.title(d)
    plt.legend()
```

```python
for d in datasets:
    for i in range(len(algs)):
        regret = [history['hypervolume_true'] for history in history_list[d][algs_name[i]]]
        std_array = [np.std(np.sort(np.array(regret)[:,t])[10:40]) for t in range(100)]
        mean_array = [np.mean(np.sort(np.array(regret)[:,t])[10:40]) for t in range(100)]
        plt.plot(mean_array, label=algs_name[i])
    plt.title(d)
    plt.legend()
```
